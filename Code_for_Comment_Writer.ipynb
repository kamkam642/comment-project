### Comment Writing Engine

#### an application to help teachers write better comments in less time


```python
# functions and code for reading data files, generating comments and writing them to files

def join_adjectives(traits:list) -> str:
    '''
    joins a list of adjectives into a string to be used in a sentence
    by kamran
    '''
    if len(traits) > 2:
        return ', '.join(traits[0:-1]) + ', and ' + traits[-1]
    elif len(traits) == 2:
        return ' and '.join(traits)
    elif len(traits) == 1:
        return traits[0]


def student_traits(positive_traits:list, negative_traits:list) -> str:
    """
    returns a sentence explaining a student's traits
    positive traits must be longer than 1
    by kamran
    """
    positive_traits = join_adjectives(positive_traits)
    negative_traits = join_adjectives(negative_traits)
    return f'you are {positive_traits}. However, you can sometimes be {negative_traits}.'


# ['first name', 'last name', 'pos traits', 'neg traits', 'Accomplishment', 'feedback', 'Strength (LO)', 'Challenges(LO)', 'Grade']


def learn_outcomes(LO):
    '''
    takes in a learning outcome and returns the description
    by Abby
    '''
    for key, value in course_dictionary.items():
        if LO == key: 
            return(value)

        
def course_description():
    '''
    prints class details 
    by Abby
    '''
    return(f"{course_dictionary['Course Name']}\n{course_dictionary['Summary']}.\n")


def formatter(line:list) -> dict:
    """
    formats line as a dictionary of traits
    By Abby and Kamran
    """
    name = (line[0], line[1])
    pos_traits = line[2].split(",")
    neg_traits = line[3].split(",")
    accomps = line[4]
    feedback = line[5]
    strong_LOs = line[6].strip().split(", ")
    weak_LOs = line[7].strip().split(", ")
    grade = line[8]
    return {'name': name,
            'pos_traits': pos_traits,
            'neg_traits': neg_traits,
            'accomplishments': accomps,
            'feedback': feedback,
            'strong_LOs': strong_LOs,
            'weak_LOs': weak_LOs,
            'grade': grade}


def write_comment(line:list):
    """
    takes a line corresponding to a student
    writes a comment using the line
    By Kamran and Abby
    """
    profile = formatter(line)
    comment = f"{' '.join(profile['name'])}\n"
    comment += f"{class_description()}\nStrengths:\n"
    for LO in profile['strong_LOs']:
        comment = comment + f"* {learn_outcomes(LO)}\n"
    comment = comment + "Challenges:\n"
    for LO in profile['weak_LOs']:
        comment = comment + f"* {learn_outcomes(LO)}\n"
    comment += f"{profile['name'][0]}, " + student_traits(profile['pos_traits'], profile['neg_traits']) + ' '
    comment += 'You ' + profile['accomplishments'] + ', great job! '
    comment += f"I would recommend you {profile['feedback']}. "
    comment += f"Your grade for this course is {profile['grade']}.\n\n"
    
    return(comment)


#the stuff below by Abby

file_name = f"{course_dictionary['Course Name']}.txt"
with open(file_name, "w") as file: 
    for student in STUDENTS:
        line = write_comment(student)
        file.write(line)

print(file_name)    

with open(file_name) as file: 
    for line in file:
        print(line.strip())
```
